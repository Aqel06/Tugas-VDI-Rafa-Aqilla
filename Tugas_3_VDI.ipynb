# Dalam methodologi, langkah pertama yang perlu dlilakukan dalam pengembangan visualisasi adalah mendefinisikan tujuan dan parameter (memperhatikan user, fungsi dan tone). Berdasarkan hal tersebut maka tugas anda adalah:

## 1. Tentukan data yang akan anda gunakan dalam visualisasi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Negara': ['Indonesia', 'Malaysia', 'Singapura', 'Thailand', 'Vietnam', 'Filipina'],
    'Tingkat_Literasi_Digital': [65, 80, 95, 70, 68, 72],
    'Penggunaan_Internet': [73, 89, 99, 78, 77, 82],
    'Kepemilikan_Smartphone': [68, 83, 97, 75, 72, 79]
}

df = pd.DataFrame(data)
print(df)
```

## 2. Berdasarkan data yang anda pilih tentukan tujuan, user, fungsi dan tone (jika ada) yang sesuai dengan visualisasi yang akan anda buat



*   Tujuan: Menggambarkan perbandingan tingkat literasi digital, penggunaan internet, dan kepemilikan smartphone di antara negara-negara ASEAN terpilih.
*   User: Pembuat kebijakan di bidang teknologi informasi dan komunikasi, peneliti, dan masyarakat umum yang tertarik dengan perkembangan digital di ASEAN.
*   Fungsi: Informatif dan komparatif, memungkinkan pengguna untuk membandingkan status digital di berbagai negara ASEAN.
*   Tone: Profesional dan objektif, menyajikan data tanpa bias. Penggunaan warna-warna netral dan desain yang bersih memberikan tone profesional.


## 3. Buatlah visualisasi dengan mengunakan library python seperti matplotlib, seaborn atau tools lain

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("Set2")

bar_width = 0.25
index = range(len(df['Negara']))

plt.bar(index, df['Tingkat_Literasi_Digital'], bar_width, label='Tingkat Literasi Digital')
plt.bar([i + bar_width for i in index], df['Penggunaan_Internet'], bar_width, label='Penggunaan Internet')
plt.bar([i + 2 * bar_width for i in index], df['Kepemilikan_Smartphone'], bar_width, label='Kepemilikan Smartphone')

plt.xlabel('Negara', fontsize=12)
plt.ylabel('Persentase (%)', fontsize=12)
plt.title('Perbandingan Tingkat Literasi Digital, Penggunaan Internet, dan Kepemilikan Smartphone di ASEAN', fontsize=14)
plt.xticks([i + bar_width for i in index], df['Negara'], rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
